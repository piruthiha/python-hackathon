# 4 Created  bins to segregate Patients among 'Mild Hypertension','Moderate ','High Hypertension','Severe Hypertension' with labels.In the given data set most of the patients comes under eithet Severe Hypertension or High Hypertension

```python
x=df['AmbulatoryVisits']
bins=[99,109,110,140,190]
labels=['Mild Hypertension','Moderate Hipertension','high Hypertension','Severe Hypertension']
x['SBP Range']=pd.cut(x['BloodPressureSystolic'],bins=bins, labels=labels, right=False)
SBP_Range=x.loc[:,['PatientMRN','BloodPressureSystolic','SBP Range']]
SBP_Range
```
